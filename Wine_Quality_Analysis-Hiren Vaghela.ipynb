# Univariate analysis (quality column)

```python
quality_counts = df['quality'].value_counts()
quality_counts
```

```python
plt.figure(figsize=(8,6))
plt.bar(quality_counts.index,quality_counts,color='pink')
plt.title("Count Plot of Quality")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.swarmplot(x="quality",y='alcohol',data=df,palette='viridis')
plt.title("Swarm Plot For Quanlity and Alcohol")
plt.xlabel("Quality")
plt.ylabel("Alcohol")
plt.show()
```

```python
df['quality'] = df['quality'].astype(str)
plt.figure(figsize=(10,8))
sns.violinplot(x="quality",y='alcohol',data=df,palette='viridis')
plt.title("Violin Plot For Quanlity and Alcohol")
plt.xlabel("Quality")
plt.ylabel("Alcohol")
plt.show()
```

```python
plt.figure(figsize=(10,8)) # line call whisker
sns.boxplot(x="quality",y='alcohol',data=df,palette='viridis')
plt.title("Box Plot For Quanlity and Alcohol")
plt.xlabel("Quality")
plt.ylabel("Alcohol")
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(df.corr(),annot=True,fmt='.2f',cmap='Pastel2',linewidths=2)
plt.title("Correlation Heatmap")
plt.show()
```

## 1 Quantity Distribution

```python
plt.figure(figsize=(7,4))
sns.countplot(x="quality",data=df,palette="viridis")
plt.title("Count Plot for quality and alcohol")
plt.show()
```

#### - Most wines fall in quality 5 and 6
#### - Very few wines have high quality (7 or above)

## 2. Alcohol vs Quality

```python
plt.figure()
plt.scatter(df['alcohol'], df['quality'])
plt.title("Alcohol vs Quality")
plt.xlabel("Alcohol")
plt.ylabel("Quality")
plt.show()
```

#### Higher alcohol content generally leads to better quality
#### Shows positive correlation

## 3. pH Distribution

```python
plt.figure()
df['pH'].plot(kind='hist', bins=20)
plt.title("pH Distribution")
plt.xlabel("pH")
plt.ylabel("Frequency")
plt.show()
```

#### Most wines have pH between 3.2 – 3.4
#### Indicates wines are slightly acidic

## 4. Quality Distribution B/W Bad and Good

```python
df['Quality_type1'] = 'bad'  # default
df.loc[df['quality'] >= 7, 'Quality_type'] = 'good'
```

```python
plt.figure(figsize=(10,6))
plt.pie(df['quality_type'].value_counts(),autopct="%1.1f%%",colors=["purple","pink"])
plt.legend(["Bad","Good"])
plt.show()
```

#### Here We have 86.5 % of Wine in Bad Categories According To Calculation So Suggestion is that Work on How to incress Good Wine Percentages.

## 5. Volatile Acidity vs Quality

```python
plt.figure()
plt.scatter(df['volatile acidity'], df['quality'])
plt.title("Volatile Acidity vs Quality")
plt.xlabel("Volatile Acidity")
plt.ylabel("Quality")
plt.show()
```

#### Higher volatile acidity → lower quality So Here We have Most of the Volatile  Acidity is medium so it Good
#### Indicates inverse relationship

## 6. Residual Sugar Distribution.

```python
plt.figure()
df['residual sugar'].plot(kind='hist', bins=20)
plt.title("Residual Sugar Distribution")
plt.xlabel("Residual Sugar")
plt.ylabel("Frequency")
plt.show()
```

#### Most wines have low sugar content
#### Few wines are sweet

## 7. Sulphates vs Quality

```python
plt.figure()
plt.scatter(df['sulphates'], df['quality'])
plt.title("Sulphates vs Quality")
plt.xlabel("Sulphates")
plt.ylabel("Quality")
plt.show()
```

#### Moderate sulphates improve wine quality
#### Too low sulphates → poor quality

## 8. Density vs Alcohol

```python
plt.figure()
plt.scatter(df['density'], df['alcohol'])
plt.title("Density vs Alcohol")
plt.xlabel("Density")
plt.ylabel("Alcohol")
plt.show()
```

#### Higher alcohol → lower density
#### Shows inverse relationship

## 9. Fixed acidity vs Density

```python
plt.figure(figsize=(8,6))
sns.distplot(df['fixed acidity'],color='purple')
plt.title("Count Plot of Quality")
# plt.xlabel("Density")
# plt.ylabel("alcohol")
plt.show()
```

#### The distribution of fixed acidity is right-skewed, with most values concentrated between 6 and 9.
#### This indicates that wines generally have moderate acidity, while very high acidity values are rare.

## 10 . Chlorides vs Quality

```python
plt.figure()
df.boxplot(column='chlorides', by='quality')
plt.title("Chlorides vs Quality")
plt.suptitle("")
plt.xlabel("Quality")
plt.ylabel("Chlorides")
plt.show()
```

#### Chloride levels tend to be higher in lower-quality wines, showing more spread and outliers.
#### Higher-quality wines generally have lower and more consistent chloride values.
